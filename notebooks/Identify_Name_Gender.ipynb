#Run gender identification

```python
novels_dir = '/content/drive/MyDrive/NLU1012_Project/novels/'

Gendered_names = pickle.load(open("/content/drive/MyDrive/NLU1012_Project/Name_results.p",'rb'))

for filename in tqdm(os.listdir(novels_dir)):
    novel_name = re.match("^[^.]+", filename).group()
    if novel_name not in Gendered_names.keys():
        print(novel_name)
        mix_gen_surname, f_names_cleaned, m_names_cleaned = identify_name_gender(filename, novels_dir)
        Gendered_names[novel_name] = {'Surname': mix_gen_surname, 'Female': f_names_cleaned, 'Male': m_names_cleaned}

pickle.dump(Gendered_names, open(r"/content/drive/MyDrive/NLU1012_Project/Name_results.p", "wb"))
```
